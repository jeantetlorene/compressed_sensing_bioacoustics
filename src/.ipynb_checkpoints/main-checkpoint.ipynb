# Preprocess .wav data to .pkl files (compression optional)

```python
preprocess = Preprocessing(
            **config.preprocessing.dict(),
            species_folder=config.data.species_folder,
            positive_class=config.data.positive_class,
            negative_class=config.data.negative_class,
        )

# Loop over train-val test split
types = ["train","test", "val"]
# Choose compression type
compression_types = ["no_compression", "ogg", "mp3", "flac"] # open issues: aac

for compression_type in compression_types:
    for dataset_type in types:
            print('Compression type:' + str(compression_type) +" - " + str(dataset_type) + ' data')
            X_calls, Y_calls, sample_rate=preprocess.create_dataset(files_path=Path(config.data.species_folder, "DataFiles", dataset_type +".txt"), dataset_type=dataset_type, preprocessing=False, data_augmentation=False, compression_type = compression_type)
            X_spectrograms = preprocess.convert_all_to_image(X_calls, sample_rate) # X_calls : segments
            X, Y = np.asarray(X_spectrograms), np.asarray(Y_calls)
            Y = preprocess._one_hot_encode(Y)
            # save X and Y in pickle files
            saving_path=Path(preprocess.species_folder, dataset_type)
            print(saving_path)
            # Save the dataset
            if compression_type == 'no_compression':
                with open(Path(saving_path, str(config.data.positive_class) + "_X_" + dataset_type + "_" + preprocess.audio_extension[1:] + "_.pkl"), "wb") as f:
                                pickle.dump(X, f)
                with open(Path(saving_path, str(config.data.positive_class) + "_Y_" + dataset_type + "_.pkl"), "wb") as f:
                                pickle.dump(Y, f)
            else: 
                with open(Path(saving_path, str(config.data.positive_class) + "_X_" + dataset_type + "_" + compression_type + ".pkl"), "wb") as f:
                                pickle.dump(X, f)
```

# Train CNNs for selected species
- Potentially adjust training epochs in model.py

```python
import pickle
import csv
import time
import sys
from pathlib import Path

def _load_dataset(species_folder, dataset_type="train", audio_extension="flac", compression_level="") -> tuple:
    """Loads the dataset from the save path"""
    
    # Construct file paths
    x_path = Path(species_folder, dataset_type, f"{config.data.positive_class}_X_{dataset_type}_{audio_extension}_{compression_level}.pkl")
    y_path = Path(species_folder, dataset_type, f"{config.data.positive_class}_Y_{dataset_type}.pkl")
    
    # Load data
    with open(x_path, "rb") as f:
        print(f"Loading: {x_path}")
        X = pickle.load(f)
    
    with open(y_path, "rb") as f:
        print(f"Loading: {y_path}")
        Y = pickle.load(f)
    
    print(f"Dataset loaded from {species_folder}/{dataset_type}")
    return X, Y

# Valid audio extensions and compression levels - Create a sublist if you dont want to run all models at once
audio_extensions = ["flac", "mp3", "ogg", "wav"]
compression_levels_flac = ["0", "6", "12"]  # For flac and ogg
compression_levels_ogg = ["0", "5", "10"]
bitrates = ["96k", "192k", "320k"]  # For mp3

# Iterate through audio extensions
for audio_extension in audio_extensions:
    if audio_extension == "flac":
        levels = compression_levels_flac
    elif audio_extension == "ogg":
        levels = compression_levels_ogg
    elif audio_extension == "mp3":
        levels = bitrates
    else:
        levels = [""]
    
    for level in levels:
        X_train, Y_train = _load_dataset(config.data.species_folder, "train", audio_extension, level)
        X_val, Y_val = _load_dataset(config.data.species_folder, "val", audio_extension, level)
        X_test, Y_test = _load_dataset(config.data.species_folder, "test", audio_extension, level)
        
        # Load training settings
        model_args = config.model.dict()
        print(model_args)

        save_path = Path("../result")  # where to save the model
        model_name = f"model_{config.data.positive_class}_{audio_extension}_{level}.pt"
        model_save_path = save_path / model_name
        
        model = Model(model_save_path, input_shape=(1, 128, X_train.shape[2]),  # X_train.shape[2] corresponds to the third dimension of X_test, which varies between species and compression
                      architecture_args=config.cnn_architecture.dict(), **model_args)

        # Train the model
        print("Training is running ...")
        train_losses, val_losses = model.train(X_train=X_train, Y_train=Y_train, X_val=X_val, Y_val=Y_val)
        print("Training finished.")
        # Get F1 score
        f1_score_train, _ = model.evaluate(X_val=X_train, Y_val=Y_train, metric="f1")
        f1_score_val, _ = model.evaluate(X_val=X_val, Y_val=Y_val, metric="f1")
        f1_score_test, _ = model.evaluate(X_val=X_test, Y_val=Y_test, metric="f1")
        trainable_params = model.get_number_of_parameters()

        print(f"F1 Score (Training): {f1_score_train}")
        print(f"F1 Score (Validation): {f1_score_val}")
        print(f"F1 Score (Test): {f1_score_test}")
        print(f"Trainable Parameters: {trainable_params}")
        
        # Save F1 scores, training time, and loss values to CSV in the model save directory
        csv_filename = save_path / model_name / f"{config.data.positive_class}_f1_{audio_extension}_{level}.csv"
        with open(csv_filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Training F1 Score", "Validation F1 Score", "Test F1 Score", "Trainable Parameters", "Train Losses", "Validation Losses"])
            writer.writerow([f1_score_train, f1_score_val, f1_score_test, trainable_params, train_losses, val_losses])
        print(f"Metrics saved to {csv_filename}")
```

# Explore F1 scores

```python
# Initialize a list to store results
results = []

# Find all CSV files in subdirectories
for csv_file in save_path.rglob("*.csv"):
    try:
        # Read CSV and extract relevant data
        df = pd.read_csv(csv_file)
        test_f1_score = df.iloc[0]["Test F1 Score"]  # Extract test F1 score
        val_f1_score = df.iloc[0]["Validation F1 Score"]  # Extract test F1 score
        train_f1_score = df.iloc[0]["Training F1 Score"]  # Extract test F1 score
        
        # Store the filename and test F1 score
        results.append({"CSV File": csv_file.name, "Test F1 Score": test_f1_score, "Validation F1 Score": val_f1_score, "Training F1 Score": train_f1_score})
    
    except Exception as e:
        print(f"Error reading {csv_file}: {e}")

# Convert results to a DataFrame and display
df_results = pd.DataFrame(results)
df_results = df_results.sort_values(by="Test F1 Score", ascending=False)  # Sort by F1 score
df_results
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
gibbon_compression = [0.973965, 0.972471, 0.972269, 0.965586, 0.965517, 0.963991, 0.963799, 0.962889, 0.958528]
gibbon_no_compression = 0.966255

ptw_compression = [0.864865, 0.860649, 0.857413, 0.848659, 0.844106, 0.828125, 0.824561, 0.763374, 0.760000]
ptw_no_compression = 0.757225

# Compute mean and standard deviation
mean_gibbon = np.mean(gibbon_compression)
std_gibbon = np.std(gibbon_compression)

mean_ptw = np.mean(ptw_compression)
std_ptw = np.std(ptw_compression)

# Plot
species = ["Gibbon", "PTW"]
compression_means = [mean_gibbon, mean_ptw]
compression_stds = [std_gibbon, std_ptw]
no_compression_scores = [gibbon_no_compression, ptw_no_compression]

x = np.arange(len(species))
width = 0.4

fig, ax = plt.subplots()
ax.bar(x - width/2, compression_means, width, yerr=compression_stds, label="Compression", capsize=5)
ax.bar(x + width/2, no_compression_scores, width, label="No Compression")

ax.set_ylabel("F1 Score")
ax.set_title("Comparison of Compression vs. No Compression F1 Scores")
ax.set_xticks(x)
ax.set_xticklabels(species)
ax.legend()

plt.show()

# Print summary
print(f"Gibbon - Compression: Mean = {mean_gibbon:.5f}, Std = {std_gibbon:.5f}")
print(f"Gibbon - No Compression: {gibbon_no_compression:.5f}")
print(f"PTW - Compression: Mean = {mean_ptw:.5f}, Std = {std_ptw:.5f}")
print(f"PTW - No Compression: {ptw_no_compression:.5f}")
```
